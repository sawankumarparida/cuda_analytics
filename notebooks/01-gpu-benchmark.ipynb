```python
import torch
import time
import pandas as pd

print("--- System Diagnostics ---")
print(f"CUDA Available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"Graphics Card: {torch.cuda.get_device_name(0)}\n")

# Create a massive 10,000 x 10,000 matrix of random numbers
size = 10000

print("--- CPU Calculation ---")
start_time = time.time()
cpu_matrix = torch.rand(size, size)
cpu_result = torch.matmul(cpu_matrix, cpu_matrix)
print(f"CPU Time: {(time.time() - start_time):.4f} seconds\n")

if torch.cuda.is_available():
    print("--- RTX 3050 Ti Calculation ---")
    start_time = time.time()
    # .cuda() moves the heavy data directly into the graphics card VRAM
    gpu_matrix = torch.rand(size, size).cuda() 
    gpu_result = torch.matmul(gpu_matrix, gpu_matrix)
    print(f"GPU Time: {(time.time() - start_time):.4f} seconds")
```
